### Data Downtime
- Periodo de tempo em que os dados estao ausentes, imprecisos ou rrados e sao apresetados em dashboards obsoletos, em relatorios inexatos e ate mesmo em tomadas inadequadas de decisao
- afetam os resultados financeiros
- 40% do tempo de equipe de dados, resolvendo downtime
- Dados imprecisos, usentes ou errados podem custar tempo, dinheiro e confiança dos clientes

### Cultura Data driven
- Comunicação sobre problema de dados
- Melhoria continua nos dados
- Confiabilidade nos dados (data trust)
- Maior foco na monetização de dados
- A nuvem facilita o gerenciamento e acesso a uma variedade maior de usuários e processamento mais rápido

- funçao de mensurar a confiabilidade, a completude e acurácia de dados em relação ao estado do que está sendo informado.

### DataOps
- processo de melhorar a confiabilidade e desempenho dos dados por meio de automação, reduzindo os silos de dados e fomentanto análises mais rápidas e tolerantes à falhas
