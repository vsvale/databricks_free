```python
spark.sql("TRUNCATE TABLE IF EXISTS dataops_prd.workspace.current_hive_tables")
for db in spark.sql("SHOW DATABASES").collect():
    db_name = db['databaseName']
    spark.sql(f"USE `{db_name}`")
    spark.sql("SHOW TABLES").select("database", "tableName").filter("tableName != '_sqldf'").write.mode("append").saveAsTable("dataops_prd.workspace.current_hive_tables")
```
